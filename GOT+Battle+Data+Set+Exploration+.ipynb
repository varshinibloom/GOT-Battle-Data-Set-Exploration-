# Which Year Had The Most Battles

```python
data['year'].value_counts()
```

# Plotting Using Bar Chart

```python

year= [299, 300, 298]
value = [20, 11, 7]
plt.bar(year, value)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Value', fontsize=10)
plt.title('Year That Had Most Number Of Battles')
plt.show()
```

# Year 299 Had The Most Battles 

# Region That Had The Most Battles

```python
region = data['region'].value_counts()
print(region)
```

# Plotting Using Bar Chart

```python

region.plot('bar')
plt.xlabel('Region', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Region That Had Most Number Of Battles')
plt.show()
```

# Riverlands has the highest battel count

# Which King lead most battles

```python

data['attacker_king'].value_counts().plot(kind='bar')
plt.xlabel('Kings', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('King That Lead More No Of Battles')
plt.show()
```

# Joffrey/Tommen Baratheon Lead More Battles

# What was the outcomes of all battles?

```python

data['attacker_outcome'].value_counts().plot(kind='bar')
plt.xlabel('Outcome', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Win / Loss')
plt.show()
```

# Different Types Of Battles

```python

data['battle_type'].value_counts().plot(kind='bar')
plt.xlabel('Battle Type', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Types Of Battle Lead')
plt.show()
```

# Different locations Of Battle

```python

data['location'].value_counts().plot(kind='bar')
plt.xlabel('Locations', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Types Of Battle Locations')
plt.show()
```
